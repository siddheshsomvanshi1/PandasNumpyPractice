```python
# MATPLOTLIB

plt.hist(df["Gender"],bins=5,color="red",edgecolor='black')
plt.title("Gender")
plt.xlabel("Gender")
plt.ylabel("Counts of Gender")
plt.grid(True)
plt.show()

group=df.groupby(["Name"])["Salary"].mean()
value=group.values #HISTOGRAM doe not support Non numerical value so we have to contvert them to numerical

plt.hist(value,bins=5,color="red",edgecolor='black')
plt.title("Gender")
plt.xlabel("Salary")
plt.ylabel("Counts of Name")
plt.grid(True)
plt.show()

```

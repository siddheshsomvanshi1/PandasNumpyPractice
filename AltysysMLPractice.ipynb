```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
# Extract (Load Data)
# Load Iris dataset
iris = load_iris(as_frame=True)

# Convert to DataFrame
df = iris.frame

# Slice to 100 rows only
df = df.iloc[:100].copy()

# Show first 5 rows
print(df.head())
```

```python
#  ETL Process
# Rename columns for clarity

# Transform (Clean and Organize Data)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Check for missing values
print(df.isnull().sum())

# Add a missing value manually for demonstration
df.loc[5, 'sepal_width'] = np.nan
df.loc[10, 'species'] = None

```

```python
# Load (Final Check)
print(df.info())
print(df.describe())
```

```python
# Preprocessing the Data
#  Handling Missing Values

# Fill numeric missing values with mean
df['sepal_width'].fillna(df['sepal_width'].mean(), inplace=True)

# Fill categorical missing with mode
df['species'].fillna(df['species'].mode()[0], inplace=True)
```

```python
#  Encode Categorical Columns
# Convert species to numeric using Label Encoding
df['species_encoded'] = df['species'].astype('category').cat.codes
```

```python
# Data Visualization
#  Using Matplotlib and Seaborn

# Histogram
plt.hist(df['sepal_length'], bins=10, color='skyblue', edgecolor='black')
plt.title("Sepal Length Distribution")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Boxplot (Outlier Detection)
sns.boxplot(x=df['sepal_width'], color='orange')
plt.title("Boxplot of Sepal Width")
plt.show()

# Scatter Plot
plt.scatter(df['sepal_length'], df['petal_length'], c='green')
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()


# Bar Plot
df['species'].value_counts().plot(kind='bar', color='purple')
plt.title("Count of Each Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

```

```python
import numpy as np
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', np.nan],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female', 'Male', 'Female'],
    'Age': [25, 30, np.nan, 45, 35, 28, 32],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Kolkata', np.nan, 'Mumbai'],
    'Salary': [50000, 60000, 55000, np.nan, 52000, 58000, 61000]
}
```

```python
df.isnull().sum()
```

```python
# df['Name'].dtype

# Filling Non-Numerical values with Unknown
df['Name']=df['Name'].fillna('Unknown',inplace=False)
print(df['Name'])
```

```python
df["Age"].dtype

df['Age']=df['Age'].fillna(df['Age'].mean())
print(df['Age'])
```

```python
# df['City'].dtype

df['City']=df['City'].fillna('Unknown')
print(df['City'])
```

```python
# df["Salary"].dtype

df['Salary']=df['Salary'].fillna(df['Salary'].mean())
print(df['Salary'])
```

```python
# Visualization using Matplotlib
import matplotlib.pyplot as plt

# Histogram Plotting
# plt.hist(df['Name'],bins=5,color='red',edgecolor='black')

# group=df.groupby("Age")['Name'].apply(list)

# Histogram of Age
df.groupby("Age")['Name'].apply(list)

plt.hist(df['Age'],bins=5,color='skyblue',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Bar chart
df.groupby('City')['Salary'].mean().plot(kind='bar', color='blue')

plt.title("Average Salary by City")
plt.xlabel("City")
plt.ylabel("Avg Salary")
plt.show()
```

```python
# Line plot of Salary

plt.plot(df['Name'], df['Salary'], marker='o')

plt.title("Salary per Person")
plt.xlabel("Name")
plt.ylabel("Salary")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
